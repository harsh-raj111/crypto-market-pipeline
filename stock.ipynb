```python
import pandas  as pd 
import requests
key = "5e5641b68b1b4624bae22e0f402b63c2"
api ="https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest"

headers = {
    "x-cmc_pro_api_key": key,
    "accept":"application/json"
}
response = requests.get(api, headers = headers)
data = response.json()
data
coins = data['data']
df = pd.json_normalize(coins)
df_clean = df[[
    "name",
    "symbol",
    "cmc_rank",
    "quote.USD.price",
    "quote.USD.market_cap",
    "quote.USD.volume_24h",
    "quote.USD.percent_change_24h",
    "quote.USD.last_updated"
]]
df_clean.columns =[
     "name",
    "symbol",
    "crypto_rank",
    "price",
    "market_cap",
    "volume_24h",
    "change_24h",
    "last_updated"
]
df_clean['last_updated'] = pd.to_datetime(df_clean['last_updated']).dt.date
df_clean.dtypes

import mysql.connector
db = mysql.connector.connect (
    host = "localhost",
    user = "root",
    password = "root",
    database = "crypto"
)
cursor = db.cursor()
sql = """
insert into crypto_prices
(name,symbol,crypto_rank,price,market_cap,volume_24h,change_24h,last_updated)
values(%s,%s,%s,%s,%s,%s,%s,%s)
"""
for index , row in df_clean.iterrows():
    values =(
   row ["name"],
   row ["symbol"],
   row ["crypto_rank"],
   row ["price"],
   row ["market_cap"],
   row ["volume_24h"],
   row["change_24h"],
   row ["last_updated"]
    )
    cursor.execute(sql,values)
    db.commit()
df_clean.to_csv('cleaned_data.csv',index=False)


```
